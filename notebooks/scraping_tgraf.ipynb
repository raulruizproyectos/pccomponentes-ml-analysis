```python
import requests
from bs4 import BeautifulSoup
import time
import os
```

```python

def obtener_urls_tarjetas():
    """PASO 1 y 2: Recorre las 17 páginas del listado y extrae TODAS las URLs"""
    url_tarjetas = []
    
    # Creamos un bucle que vaya desde la página 1 hasta la 17
    for pagina in range(1, 18):
        lista_url = f"https://www.pccomponentes.com/tarjetas-graficas?page={pagina}"

        response = requests.get(lista_url)
        soup = BeautifulSoup(response.text, 'html.parser')
        
        tarjetas_graficas = soup.find_all("a", attrs={"data-testid": "normal-link"})

        for tarjeta in tarjetas_graficas:
            url = tarjeta['href']
            if "tarjeta-grafica" in url and len(url) > 70:
                url_tarjetas.append(url)
                
        # Imprimimos un aviso para saber que va avanzando bien por las páginas
        print(f"Página {pagina} procesada de forma correcta.")
            
    return url_tarjetas
```

```python
def descargar_archivos_html(url_tarjetas):
    """PASO 3: Recorre la lista de URLs y descarga los HTMLs"""
    for url in url_tarjetas:
        nombre_limpio = url.split("/")[-1]
        ruta_archivo = f"data_raw_tarjetas/{nombre_limpio}.html"

        response = requests.get(url=url)
        with open(ruta_archivo, "w", encoding="utf-8") as archivo:
            archivo.write(response.text)

        print(f"Guardado: {nombre_limpio} | Estado: {response.status_code}")

        time.sleep(2)
```
