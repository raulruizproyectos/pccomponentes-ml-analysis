```python
# Importamos la librería psycopg para conectar con PostgreSQL

import psycopg

print("psycopg importado correctamente")
```

```python
# Establecemos la conexión a la base de datos PostgreSQL utilizando la URI creada

conn = psycopg.connect(DATABASE_URI)

print("Conexion correcta")
```

```python
# Creamos un cursor para ejecutar consultas SQL

cursor = conn.cursor()

print("Cursor creado correctamente")
```

```python
# Ejecutamos una consulta SQL para obtener los datos de la tabla "productos"

cursor.execute("""
    SELECT producto_id, modelo, precio
    FROM productos;
""")
```

```python
# Obtenemos los resultados de la consulta y los almacenamos en una variable

productos = cursor.fetchall()

productos
```

```python
# Mostramos los datos obtenidos de la consulta SQL

for producto in productos:
    producto_id = producto[0]
    modelo = producto[1]
    precio = producto[2]

    print(f"ID: {producto_id}")
    print(f"Modelo: {modelo}")
    print(f"Precio: {precio} euros")
```

```python
# Ejecutamos una consulta SQL para obtener los datos de la tabla "productos" junto con sus especificaciones, aqui no se muestran los datos.

cursor.execute("""
    SELECT
        p.producto_id,
        p.modelo,
        p.precio,
        e.gpu,
        e.memoria_vram,
        e.tipo_memoria,
        e.bus_memoria
    FROM productos p
    JOIN especificaciones_producto e
        ON p.producto_id = e.producto_id;
""")
```

```python
# Obtenemos los resultados de la consulta y los almacenamos en una variable

productos_especificaciones = cursor.fetchall()

productos_especificaciones
```

```python
# Mostramos los datos obtenidos de la consulta SQL

for producto in productos_especificaciones:
    producto_id = producto[0]
    modelo = producto[1]
    precio = producto[2]
    gpu = producto[3]
    memoria_vram = producto[4]
    tipo_memoria = producto[5]
    bus_memoria = producto[6]

    print(f"ID: {producto_id}")
    print(f"Modelo: {modelo}")
    print(f"Precio: {precio} euros")
    print(f"GPU: {gpu}")
    print(f"Memoria VRAM: {memoria_vram}")
    print(f"Tipo de memoria: {tipo_memoria}")
    print(f"Bus de memoria: {bus_memoria}")
```

```python
# Ejecutamos una consulta SQL para obtener los datos de la tabla "productos" junto con sus reseñas, aqui no se muestran los datos.

cursor.execute("""
    SELECT
        p.producto_id,
        p.modelo,
        r.usuario,
        r.valoracion,
        r.texto_resena,
        r.pros,
        r.contras
    FROM productos p
    JOIN resenas r
        ON p.producto_id = r.producto_id;
""")
```

```python
# Obtenemos los resultados de la consulta y los almacenamos en una variable

productos_resenas = cursor.fetchall()

productos_resenas
```

```python
# Mostramos los datos obtenidos de la consulta SQL

for resena in productos_resenas:
    producto_id = resena[0]
    modelo = resena[1]
    usuario = resena[2]
    valoracion = resena[3]
    texto_resena = resena[4]
    pros = resena[5]
    contras = resena[6]

    print(f"Producto ID: {producto_id}")
    print(f"Modelo: {modelo}")
    print(f"Usuario: {usuario}")
    print(f"Valoracion: {valoracion}")
    print(f"Texto: {texto_resena}")
    print(f"Pros: {pros}")
    print(f"Contras: {contras}")
```
