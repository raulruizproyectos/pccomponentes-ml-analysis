# AnÃ¡lisis de sentimiento de reseÃ±as RAM

Este notebook utiliza el modelo `nlptown/bert-base-multilingual-uncased-sentiment`
para clasificar el sentimiento de las reseÃ±as segÃºn su texto.

```python
from transformers import pipeline

MODELO_SENTIMIENTO = "nlptown/bert-base-multilingual-uncased-sentiment"

print(f"Modelo seleccionado: {MODELO_SENTIMIENTO}")
```

```python
analizador_sentimiento = pipeline(
    "text-classification",
    model=MODELO_SENTIMIENTO,
    device=-1,
)

print("Modelo cargado correctamente en la CPU")
```

## Prueba inicial del modelo

Probamos el modelo con tres reseÃ±as inventadas para comprobar que interpreta
correctamente sentimientos positivos, neutrales y negativos.

```python
resenas_prueba = [
    "Excelente memoria RAM, muy rÃ¡pida y funciona perfectamente.",
    "Cumple su funciÃ³n, pero no destaca especialmente.",
    "Producto defectuoso, dejÃ³ de funcionar al segundo dÃ­a.",
]

resultados_prueba = analizador_sentimiento(resenas_prueba)

for resena, resultado in zip(resenas_prueba, resultados_prueba):
    print(resena)
    print(resultado)
    print()
```

## InterpretaciÃ³n del resultado

Convertimos las estrellas predichas por el modelo en tres categorÃ­as:

- 1 o 2 estrellas: negativo.
- 3 estrellas: neutral.
- 4 o 5 estrellas: positivo.

```python
SENTIMIENTO_POR_ESTRELLAS = {
    1: "negativo",
    2: "negativo",
    3: "neutral",
    4: "positivo",
    5: "positivo",
}


def interpretar_resultado(resultado):
    estrellas = int(resultado["label"].split()[0])

    return {
        "estrellas_predichas": estrellas,
        "sentimiento": SENTIMIENTO_POR_ESTRELLAS[estrellas],
        "confianza": resultado["score"],
    }


resultados_interpretados = [
    interpretar_resultado(resultado)
    for resultado in resultados_prueba
]

resultados_interpretados
```

## ConexiÃ³n con PostgreSQL

Creamos una conexiÃ³n de solo lectura para obtener una pequeÃ±a muestra de
reseÃ±as RAM reales. La contraseÃ±a no se guarda en el notebook.

```python
from getpass import getpass

import psycopg


password = getpass("ContraseÃ±a PostgreSQL: ")

conn = psycopg.connect(
    dbname="pccomponentes_ml",
    user="postgres",
    password=password,
    host="localhost",
    port=5432,
)

conn.execute("SET TRANSACTION READ ONLY")

print("ConexiÃ³n de solo lectura creada correctamente")
```

## Muestra de reseÃ±as reales

Seleccionamos diez reseÃ±as de cada grupo de valoraciÃ³n para comparar las
estrellas del usuario con la predicciÃ³n obtenida a partir del texto.

```python
import pandas as pd


consulta_muestra = """
WITH resenas_clasificadas AS (
    SELECT
        r.resena_id,
        r.valoracion,
        r.texto_resena,
        CASE
            WHEN r.valoracion < 3 THEN 'negativo'
            WHEN r.valoracion = 3 THEN 'neutral'
            ELSE 'positivo'
        END AS sentimiento_estrellas,
        ROW_NUMBER() OVER (
            PARTITION BY
                CASE
                    WHEN r.valoracion < 3 THEN 'negativo'
                    WHEN r.valoracion = 3 THEN 'neutral'
                    ELSE 'positivo'
                END
            ORDER BY RANDOM()
        ) AS numero
    FROM resenas AS r
    JOIN productos AS p
        ON p.producto_id = r.producto_id
    WHERE p.categoria = 'memoria_ram'
      AND TRIM(r.texto_resena) <> ''
)
SELECT
    resena_id,
    valoracion,
    texto_resena,
    sentimiento_estrellas
FROM resenas_clasificadas
WHERE numero <= 10
ORDER BY valoracion;
"""

with conn.cursor() as cursor:
    cursor.execute(consulta_muestra)
    filas = cursor.fetchall()
    columnas = [columna.name for columna in cursor.description]

muestra_resenas = pd.DataFrame(filas, columns=columnas)

muestra_resenas["sentimiento_estrellas"].value_counts()
```

## ValidaciÃ³n con reseÃ±as reales

Aplicamos el modelo a una muestra equilibrada y comparamos el sentimiento
predicho desde el texto con el sentimiento calculado mediante las estrellas.

```python
predicciones_muestra = analizador_sentimiento(
    muestra_resenas["texto_resena"].tolist(),
    truncation=True,
    max_length=512,
)

predicciones_muestra = [interpretar_resultado(resultado) for resultado in predicciones_muestra]

muestra_resenas["estrellas_predichas"] = [resultado["estrellas_predichas"] for resultado in predicciones_muestra]

muestra_resenas["sentimiento_modelo"] = [resultado["sentimiento"] for resultado in predicciones_muestra]

muestra_resenas["confianza"] = [resultado["confianza"] for resultado in predicciones_muestra]

muestra_resenas["coincide"] = (muestra_resenas["sentimiento_estrellas"] == muestra_resenas["sentimiento_modelo"])

muestra_resenas[
    [
        "valoracion",
        "sentimiento_estrellas",
        "estrellas_predichas",
        "sentimiento_modelo",
        "confianza",
        "coincide",
    ]
]
```

## Resultado de la validaciÃ³n

Calculamos el porcentaje de coincidencia entre el sentimiento obtenido desde
las estrellas y el sentimiento predicho mediante el texto.

```python
resumen_validacion = (
    muestra_resenas
    .groupby("sentimiento_estrellas")["coincide"]
    .agg(aciertos="sum", total="count")
)

resumen_validacion["porcentaje"] = (
    resumen_validacion["aciertos"]
    / resumen_validacion["total"]
    * 100
).round(1)

resumen_validacion
```

## RevisiÃ³n de diferencias

Revisamos las reseÃ±as donde el sentimiento del texto no coincide con la
clasificaciÃ³n obtenida mediante las estrellas.

```python
for _, fila in muestra_resenas[~muestra_resenas["coincide"]].iterrows():
    print(f"ValoraciÃ³n original: {fila['valoracion']}")
    print(f"Sentimiento del modelo: {fila['sentimiento_modelo']}")
    print(f"Confianza: {fila['confianza']:.2%}")
    print(f"Texto: {fila['texto_resena']}")
    print("-" * 80)
```

## Procesamiento completo de reseÃ±as RAM

Una vez validado el modelo con una muestra pequeÃ±a, se procesan todas las reseÃ±as de memoria RAM que todavÃ­a no tienen resultado guardado.

Los resultados se almacenan en PostgreSQL para no ejecutar el modelo cada vez que se consulte la aplicaciÃ³n.

```python
consulta_resenas_pendientes = """
SELECT
    r.resena_id,
    r.texto_resena
FROM resenas AS r
JOIN productos AS p
    ON p.producto_id = r.producto_id
LEFT JOIN resultados_sentimiento AS rs
    ON rs.resena_id = r.resena_id
WHERE p.categoria = 'memoria_ram'
  AND TRIM(r.texto_resena) <> ''
  AND rs.resena_id IS NULL
ORDER BY r.resena_id;
"""

with conn.cursor() as cursor:
    cursor.execute(consulta_resenas_pendientes)
    filas = cursor.fetchall()
    columnas = [columna.name for columna in cursor.description]

resenas_pendientes = pd.DataFrame(filas, columns=columnas)

consulta_insertar_sentimiento = """
INSERT INTO resultados_sentimiento (
    resena_id,
    estrellas_predichas,
    sentimiento,
    confianza,
    modelo
)
VALUES (%s, %s, %s, %s, %s)
ON CONFLICT (resena_id) DO UPDATE SET
    estrellas_predichas = EXCLUDED.estrellas_predichas,
    sentimiento = EXCLUDED.sentimiento,
    confianza = EXCLUDED.confianza,
    modelo = EXCLUDED.modelo;
"""

conn_escritura = psycopg.connect(
    dbname="pccomponentes_ml",
    user="postgres",
    password=password,
    host="localhost",
    port=5432,
)

tamano_lote = 32
total_guardado = 0

for inicio in range(0, len(resenas_pendientes), tamano_lote):
    lote = resenas_pendientes.iloc[inicio:inicio + tamano_lote]

    predicciones = analizador_sentimiento(
        lote["texto_resena"].tolist(),
        truncation=True,
        max_length=512,
    )

    predicciones = [interpretar_resultado(resultado) for resultado in predicciones]

    filas_insertar = [
        (
            resena_id,
            int(resultado["estrellas_predichas"]),
            resultado["sentimiento"],
            float(resultado["confianza"]),
            MODELO_SENTIMIENTO,
        )
        for resena_id, resultado in zip(lote["resena_id"], predicciones)
    ]

    with conn_escritura.cursor() as cursor:
        cursor.executemany(consulta_insertar_sentimiento, filas_insertar)

    conn_escritura.commit()
    total_guardado += len(filas_insertar)

    print(f"Guardadas {total_guardado} de {len(resenas_pendientes)} reseÃ±as pendientes")

conn_escritura.close()

print(f"Proceso terminado. Resultados nuevos guardados: {total_guardado}")
```
