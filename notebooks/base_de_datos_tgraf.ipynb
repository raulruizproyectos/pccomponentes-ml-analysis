```python
import psycopg
def volcado_datos(dsn, lista_productos, lista_especificaciones, lista_distribucion, lista_resenas):
    with psycopg.connect(dsn) as conn:
        with conn.cursor() as cursor:

            # 1. Tabla Productos
            query = "INSERT INTO productos (id_producto, url, categoria, modelo, marca, precio, valoracion_media, numero_total_opiniones, porcentaje_recomendacion, numero_recomendaciones) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s);"
            cursor.executemany(query=query, params_seq=lista_productos)

            # 2. Tabla Especificaciones GPU
            query = "INSERT INTO especificaciones_gpu (id_producto, gpu, memoria_vram, tipo_memoria, bus_memoria, velocidad_memoria, reloj_base, reloj_boost, salidas_video, resolucion_maxima) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s);"
            cursor.executemany(query=query, params_seq=lista_especificaciones)

            # 3. Tabla Distribución Valoraciones
            query = "INSERT INTO distribucion_valoraciones (id_producto, estrellas_5, estrellas_4, estrellas_3, estrellas_2, estrellas_1, fuente_desglose) VALUES (%s, %s, %s, %s, %s, %s, %s);"
            cursor.executemany(query=query, params_seq=lista_distribucion)

            # 4. Tabla Reseñas
            query = "INSERT INTO resenas (id_producto, valoracion, fecha_resena_texto, variante_modelo, texto_resena, pros, contras, numero_total_opiniones) VALUES (%s, %s, %s, %s, %s, %s, %s, %s);"
            cursor.executemany(query=query, params_seq=lista_resenas)
```
